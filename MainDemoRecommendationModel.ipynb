```python
import numpy as np
import tensorflow as tf
# tf.enable_eager_excecution()
print(tf.__version__)
```

```python
import pandas as pd
Customerdata = pd.read_excel('/content/unmerged_tmp.xlsx')
Productcategory = pd.read_excel('/content/Product_category.xlsx')
```

```python
num_recommendations = 2

Member = list(Customerdata['member_id'])
Unique_Member = list(set(Member))

Productname = list(Customerdata['Product Name'])
Unique_Productname = list(set(Productname))

ProductCategory = list(Customerdata['Product Category'])
Unique_ProductCategory = list(set(ProductCategory))

Quantity = list(Customerdata['Total Quantity '])

num_mem = len(Unique_Member)
num_Product = len(Unique_Productname)
num_cat = len(Unique_ProductCategory)

print(len(Unique_Productname))
print(len(Unique_Member))
```

```python
Customerdata["Total Quantity "]
```

```python
member_product = []
Quantity=0
for eachmember in Unique_Member:
    mem_list = []

    for eachproduct in Unique_Productname:
       Quantity=0
       for ind in Customerdata.index:
          if eachmember == Customerdata["member_id"][ind] and eachproduct == Customerdata["Product Name"][ind]:
            #print(eachmember)
            #print(eachproduct)
            Quantity = Customerdata["Total Quantity "][ind]
           # print(Quantity)

       mem_list.append(Quantity)
    member_product.append(mem_list)
tf. cast(member_product, tf. int32)
print(member_product)
```

```python
Product_Categoey = []
occur=0
for eachproduct in Unique_Productname:
      product_list = []

      for eachproductcat in Unique_ProductCategory:
        occur=0
        for ind in Productcategory.index:

            if eachproductcat == Productcategory["Product Category"][ind] and eachproduct == Productcategory["Product Name"][ind]:
              #print(eachmember)
              #print(eachproduct)
               occur = 1
            # print(Quantity)

        product_list.append(occur)
      Product_Categoey.append(product_list)
tf. cast(Product_Categoey, tf. float32)
print(Product_Categoey)
```

```python
usercatergory = tf.matmul(member_product,Product_Categoey)
tf. cast(usercatergory, tf. float32)
usercatergory
```

```python
usercatergory = usercatergory/tf.reduce_sum(usercatergory,axis =1 ,keepdims = True)
usercatergory
```

```python
top_user_cat =tf.nn.top_k(usercatergory,num_cat)[1]
top_user_cat
```

```python
for i in range(num_mem):
  cat_names = [Unique_ProductCategory[int(index)] for index in top_user_cat[i]]
  print('{}:{}'.format(Unique_Member[i],cat_names))
```

```python
users_Quantity = tf.matmul(usercatergory,tf.transpose(Product_Categoey))
tf. cast(users_Quantity, tf. int32)
users_Quantity
```

```python
user_Quantity_new = tf.where(tf.equal(member_product,tf.zeros_like(member_product)),
                           users_Quantity,
                           tf.zeros_like(tf.cast(member_product,tf.int32)))
user_Quantity_new
```

```python
top_product = tf.nn.top_k(user_Quantity_new,num_recommendations)[1]
top_product
```

```python
for i in range(num_mem):
  product_names = [Unique_Productname[int(index)] for index in top_product[i]]
  print('{}:{}'.format(Unique_Member[i],product_names))
```
