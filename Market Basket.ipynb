```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
# pd.set_option('max_rows', None)
# pd.set_option('max_columns', None)
```

```python
df = pd.read_excel("/content/processed product list.xlsx")
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
df.describe()
```

```python
df['bill_detail[0].product_name'] = df['bill_detail[0].product_name'].str.strip()
df['transaction_no'] = df['transaction_no'].astype('str')
```

```python
basket = (df.groupby(['transaction_no', 'bill_detail[0].product_name'])['bill_detail[0].qty'] #group them on InvoiceNo and Description based on Quantity
          .sum().unstack().reset_index().fillna(0) #sum the quantity, unstack them, fill 0 to nan values.
          .set_index('transaction_no')) #set the index as Invoice Number
```

```python
basket.head()
```

```python
basket.shape
```

```python
#below function converts a values < 0 to 0 and value greater than equal 1 to 1.
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
#apply function to data using applymap.
basket_sets = basket.applymap(encode_units)
```

```python
basket_sets.head()
```

```python
frequent_itemsets = apriori(basket_sets, min_support=0.01, use_colnames=True)
```

```python
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
```

```python
rules[(rules['lift'] >= 1) &
       (rules['confidence'] >= 0.1)]
```
